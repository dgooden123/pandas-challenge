```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os

# File to Load (Remember to Change These)
school_data_path = "Resources/schools_complete.csv"
student_data_path = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_path)
student_data = pd.read_csv(student_data_path)

#Preview data
#school_data.head()
#student_data.head()



```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#find names of schools
schools=school_data["school_name"].unique()

#find the number of schools
total_schools=len(schools)
total_schools

#calculate the number of students by finding the sum of each school's size
total_students=school_data["size"].sum()

#calculate the total budget using .sum
total_budget=school_data["budget"].sum()

#calculate average math score
avg_math=student_data["math_score"].mean()

#calculate average reading score
avg_reading=student_data["reading_score"].mean()

#use locate to find students with a passsing math score
math_pass_n=student_data.loc[student_data["math_score"]>=70]["math_score"].count()

#percent pass
math_pass_p=math_pass_n/total_students
math_pass_p="{:.2%}".format(math_pass_p)
#math_pass_p

#use locate to find students with a passsing reading score
reading_pass_n=student_data.loc[student_data["reading_score"]>=70]["reading_score"].count()

#percent pass
reading_pass_p=reading_pass_n/total_students
reading_pass_p="{:.2%}".format(reading_pass_p)


#find students who passed both
both_pass_n=student_data.loc[(student_data['math_score'] >= 70) & 
                             (student_data['reading_score'] >= 70)]['student_name'].count()
#percent pass both
both_pass_p=both_pass_n/total_students
both_pass_p="{:.2%}".format(both_pass_p)


#create dataframe to hold new variables
district_table = pd.DataFrame({
    
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Reading Score": [avg_reading],
    "Average Math Score": [avg_math],
    "% Passing Reading":[reading_pass_p],
    "% Passing Math": [math_pass_p],
    "Overall Passing Rate": [both_pass_p]

})
district_table
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Combine the data into a single dataset and retain all students  
complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])


# School Name 
school_name_group = complete.groupby(['school_name'])


# school types using school dataset
school_type = school_data.set_index('school_name')['type']


# # total students by school, use student ID and gender to make sure values are the same
#students_by_school = school_name_group['gender'].count()
students_by_school = school_name_group['Student ID'].count()

 

# Total school budget using the school dataset
school_budget= school_data.set_index('school_name')['budget']

# Budget per student by school
school_budget_student=school_budget/students_by_school

# Average math score by school 
avg_math_school=school_name_group['math_score'].sum()/students_by_school

# Average reading score by school 
avg_read_school=school_name_group['reading_score'].sum()/students_by_school

# % passing math by school
pass_math_school = complete[complete['math_score'] >= 70].groupby('school_name')['Student ID'].count()/students_by_school 


# % passing reading by school
pass_read_school = complete[complete['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/students_by_school


# % passing both reading and math by school
pass_both_school = complete[(complete['reading_score'] >= 70) & 
                    (complete['math_score'] >= 70)].groupby('school_name')['Student ID'].count()/students_by_school


school_table = pd.DataFrame({
    "School Type": school_type,
    "Total Students": students_by_school,
    "Total School Budget": school_budget,
    "Per Student Budget": school_budget_student,
    "Average Math Score": avg_math_school,
    "Average Reading Score": avg_read_school,
    '% Passing Math': pass_math_school,
    '% Passing Reading': pass_read_school,
    "% Overall Passing": pass_both_school
})
school_table
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
#sort the school_summary table by the overall passing rate
top_performers=school_table.sort_values("% Overall Passing", ascending=False)

#use .head to automatically print the top 5
top_performers.head()
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
#sort the school_summary table by the overall passing rate
low_performers=school_table.sort_values("% Overall Passing", ascending=True)

#use .head to automatically print the top 5
low_performers.head()
```

## Math Scores by Grade

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#create bins using the budget per student, range is 578-655
bins = [0, 584, 630, 643.999, 700]
group_name = ['< $584', "$585 - $629", "$630 - $644", "$645 - $675"]
complete['bins'] = pd.cut(complete['budget']/complete['size'], bins, labels = group_name)

#groupby newly created bins
spending_groups = complete.groupby('bins')

#calculate average math score by spending group
avg_math = spending_groups['math_score'].mean()

#calculate average reading score by spending group
avg_read = spending_groups['reading_score'].mean()

#calculate %passing math
math_pass = complete[complete['math_score'] >= 70].groupby('bins')['Student ID'].count()/spending_groups['Student ID'].count()

#calculate %passing reading
reading_pass = complete[complete['reading_score'] >= 70].groupby('bins')['Student ID'].count()/spending_groups['Student ID'].count()

#calculate %passing both
both_pass = complete[(complete['reading_score'] >= 70) & (complete['math_score'] >= 70)].groupby('bins')['Student ID'].count()/spending_groups['Student ID'].count()

            
# scores by spending table 

scores_by_spending = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': math_pass,
    '% Passing Reading': reading_pass,
    "Overall Passing Rate": both_pass
    })
scores_by_spending.index.name = "Spending Ranges (Per Student)"


scores_by_spending  
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
bins = [0, 999, 1999, 99999999999]
group_name = ["Small (<1000)", "Medium (1000-2000)" , "Large (>2000)"]
complete['size bins'] = pd.cut(complete['size'], bins, labels = group_name)

#groupby newly created bins
size_groups = complete.groupby('size bins')

#calculate average math score by size group
avg_math = size_groups['math_score'].mean()

#calculate average reading score by size group
avg_read = size_groups['reading_score'].mean()

#calculate %passing math
math_pass = complete[complete['math_score'] >= 70].groupby('size bins')['Student ID'].count()/size_groups['Student ID'].count()

#calculate %passing reading
reading_pass = complete[complete['reading_score'] >= 70].groupby('size bins')['Student ID'].count()/size_groups['Student ID'].count()

#calculate %passing both
both_pass = complete[(complete['reading_score'] >= 70) & (complete['math_score'] >= 70)].groupby('size bins')['Student ID'].count()/size_groups['Student ID'].count()

            
# scores by size table 

scores_by_size = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': math_pass,
    '% Passing Reading': reading_pass,
    "Overall Passing Rate": both_pass
    })

scores_by_size.index.name = "School Size"


scores_by_size  
```
